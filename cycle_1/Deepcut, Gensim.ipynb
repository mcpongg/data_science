```python
from gensim.models import Word2Vec
import deepcut
```

```python
# ชุดติดตั้งหลังคาระบบแห้ง , เชิงชายคอนวูด

list_queries = ['ชุดติดตั้งหลังคาระบบแห้ง','เชิงชายคอนวูด']
list_word = ['เชิงชาย', 'คอนวูด', 'ติดตั้ง']

for query in list_queries:
    
    custom_dict = deepcut.tokenize(query , custom_dict= list_word)
    no_dict =  deepcut.tokenize(query)

    print ('no dict:', no_dict)

    print ('custom dict: ', custom_dict, '\n')
```

```python
model = Word2Vec.load("1_word2vec_model/query_window1.model")
```

```python
model.wv.index2word
```

```python
text = 'เหล็ก'
vec = model.wv.word_vec(text)

print('query: ', text)
vec
```
